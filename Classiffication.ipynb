```python
def style_telecom_churn_data(styler, df):
    
    styler.background_gradient(
        subset=['weekly_mins_watched'],
        cmap='Greens',
        vmax=df['weekly_mins_watched'].max(),
        vmin=df['weekly_mins_watched'].min()
    )
    styler.background_gradient(
        subset=['no_of_days_subscribed'],
        cmap='Oranges',
        vmax=df['no_of_days_subscribed'].max(),
        vmin=df['no_of_days_subscribed'].min()
    )
    styler.background_gradient(
        subset=['churn'],
        cmap='Reds',
        vmin=0, 
        vmax=1
    )
    
    styler.format(precision=2)
    
    styler.set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#f0f0f0'), ('color', '#333333'), ('font-weight', 'bold')]},
        {'selector': 'td', 'props': [('border', '1px solid #e0e0e0')]},
        {'selector': 'table', 'props': [('border-collapse', 'collapse'), ('width', '100%')]}
    ])
    styler.set_caption("Telecom Customer Data Sample with Churn Indicators")
    return styler
```
